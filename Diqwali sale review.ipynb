```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv',encoding = 'unicode_escape')
df.head()
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.columns
```

```python
df.rename(columns = {'marital_status' : 'merried_type'})
```

```python
df.describe()
```

```python
ax = sns.countplot(x = 'Gender',data = df)
plt.show()
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

```python
#In Above Graph We can see that Group of (26 - 35) Has Done Highest amount of sales.
```

```python
df.columns
```

```python

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(15)

sns.set(rc={'figure.figsize':(17,10)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

```python
#From above graph we can see that Uttarpradesh has highest sell ratio.
```

```python
correlation = df['Age'].corr(df['Amount'])
print(correlation)
```

# Conclusion
#Age of woman between(26 - 35) From Uttar Pradesh ,Maharastra, Karnataka are more likely to buy product. Group of age (0-16) from punjab has least number of buy products.
